## Quality Issues
- Duplicated records.
- Branch column is repeated column.
- Nulls represented as dashes (-) in Customer type column.
- Input mistake (Memberr in Customer type column).
- Unit price has USD sometimes.
- Unit price is an object not a float.
- Quantity column has negative values (-1, -7, -8, -8) instead of (1, 7, 8, 8).
- Missing values in Tax 5% column.
- Missing values in Total column.
- Dates in different formats.
- Times in different formats.
- Invoice 307-85-2293 rating is (97) instead of (9.7).


## Tidiness Issues
- Columns headers as values not variables (Columns: Yangon, Naypyitaw, Mandalay)

# Clean

```python
Supermarket_Sales_Clean = Supermarket_Sales.copy()
```

## Duplicated Data

### Define:
Remove Duplicate Records.

### Code:

```python
Supermarket_Sales_Clean.drop_duplicates(inplace=True)
```

### Test:

```python
Supermarket_Sales_Clean.info()
```

## Data Redundancy

### Define:
Drop Branch Column

### Code:

```python
Supermarket_Sales_Clean.drop('Branch', axis=1, inplace=True)
```

### Test:

```python
Supermarket_Sales_Clean.info()
```

## Inaccurate Data

### Define:
Correct input mistake (Change Memberr into Member in Customer type column)

### Code:

```python
Supermarket_Sales_Clean['Customer type'] = Supermarket_Sales_Clean['Customer type'].str.replace('Memberr', 'Member')
```

### Test:

```python
Supermarket_Sales_Clean['Customer type'].unique()
```

### Define:
Correct input mistake in Rating Column

### Code:

```python
Supermarket_Sales_Clean['Rating'] = np.where(Supermarket_Sales_Clean['Rating'] >10, Supermarket_Sales_Clean['Rating'] * 0.1, Supermarket_Sales_Clean['Rating'])
Supermarket_Sales_Clean['Rating'] = Supermarket_Sales_Clean['Rating'].round(1)
```

### Test:

```python
Supermarket_Sales_Clean[Supermarket_Sales_Clean['Rating'] > 10]
```

## Invalid Data

### Define:
Remove USD from Unit price column. Change Unit Price to float.

### Code:

```python
Supermarket_Sales_Clean['Unit price'] = Supermarket_Sales_Clean['Unit price'].str.replace('USD', '')
Supermarket_Sales_Clean['Unit price'] = Supermarket_Sales_Clean['Unit price'].astype(float)
```

### Test:

```python
Supermarket_Sales_Clean.info()
```

```python
Supermarket_Sales_Clean.sample(10)
```

### Define:
Change Quantity column negative values.

### Code:

```python
Supermarket_Sales_Clean['Quantity'] = Supermarket_Sales_Clean['Quantity'].abs()
```

### Test:

```python
Supermarket_Sales_Clean[Supermarket_Sales_Clean['Quantity'] < 0]
```

## Missing Data

### Define:
Recalcuate the Tax 5% column: Unit price multiply Quantity multiply 0.05

### Code:

```python
Supermarket_Sales_Clean['Tax 5%'] = Supermarket_Sales_Clean['Unit price'] * Supermarket_Sales_Clean['Quantity'] * 0.05
Supermarket_Sales_Clean['Tax 5%'] = Supermarket_Sales_Clean['Tax 5%'].round(4)
```

### Test:

```python
Supermarket_Sales_Clean.info()
```

```python
Supermarket_Sales_Clean.sample(10)
```

### Define:
Recalculate the Total column: Unit price multiply Quantity plus Tax 5%

### Code:


```python
Supermarket_Sales_Clean['Total'] = Supermarket_Sales_Clean['Unit price'] * Supermarket_Sales_Clean['Quantity'] + Supermarket_Sales_Clean['Tax 5%']
Supermarket_Sales_Clean['Total'] = Supermarket_Sales_Clean['Total'].round(4)
```

### Test:

```python
Supermarket_Sales_Clean.info()
```

```python
Supermarket_Sales_Clean.sample(10)
```

## Inconsistent Data

### Define:
Change Date column format

### Code:

```python
Supermarket_Sales_Clean['Date'] = pd.to_datetime(Supermarket_Sales_Clean['Date'])
```

### Test:

```python
Supermarket_Sales_Clean.sample(10)
```

### Define:
Change Time column format

### Code:

```python
def Time_Format(t):
  if 'PM' in t or 'AM' in t:
    return pd.to_datetime(t, format='%I - %M %p').strftime('%H:%M')
  else:
    return pd.to_datetime(t, format='%H:%M').strftime('%H:%M')

Supermarket_Sales_Clean['Time'] = Supermarket_Sales_Clean['Time'].apply(Time_Format)
```

### Test:

```python
Supermarket_Sales_Clean.sample(10)
```

## Tidiness Issues

### Define:
Convert the Yangon, Naypyitaw and Mandalay Columns to Branch location. Drop the Yangon, Naypyitaw and Mandalay Columns

### Code:

```python
def get_Branch_location(row):
  if row['Yangon'] == 1:
    return 'Yangon'
  elif row['Naypyitaw'] == 1:
    return 'Naypyitaw'
  elif row['Mandalay'] == 1:
    return 'Mandalay'


Supermarket_Sales_Clean['Branch location'] = Supermarket_Sales_Clean.apply(get_Branch_location, axis=1)
Supermarket_Sales_Clean.drop(['Yangon', 'Naypyitaw', 'Mandalay'], axis=1, inplace=True)
```

### Test:

```python
Supermarket_Sales_Clean.info()
```

Rating Distribution

```python
def Rating_Distribution(row):
  if 0 <= row['Rating'] < 4:
    return 'Bad'
  elif 4 <= row['Rating'] < 6:
    return 'Normal'
  elif 6 <= row['Rating'] < 8:
    return 'Good'
  elif 8 <= row['Rating'] <= 10:
    return 'Excellent'

Rating_Counts = Supermarket_Sales_Clean.apply(Rating_Distribution, axis=1).value_counts()

plt.figure(figsize=(8, 6))
plt.pie(Rating_Counts, labels=Rating_Counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Rating Distribution')
plt.show()

```

Product Line Rating

```python
Supermarket_Sales_Clean['Rating_Distribution'] = Supermarket_Sales_Clean.apply(Rating_Distribution, axis=1)

Product_Line_Rating = Supermarket_Sales_Clean.groupby(['Product line', 'Rating_Distribution']).size().unstack()

Product_Line_Rating.plot(kind='bar', figsize=(14, 8))
plt.title('Product Line Rating')
plt.xlabel('Product Line')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Rating')
plt.show()
```
